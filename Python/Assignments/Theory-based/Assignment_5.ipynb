### **_3. What is the most significant distinction between a dictionary and a list?_**

**_Dictionary is the hashed structure of key value pairs. <br>
It consists of key value pairs where elements are accessed by the key associated with it._**

```python
d = {'a' : 1, 'b' : 2, 'c' : 3}
```

```python
d.keys()
```

```python
d.values()
```

**_Lists are the indexed based data structure.<br>
It consists of indices where elements are stored at a certain index location._**

```python
l = [1,2,3,4,5]
```

```python
l[0]
```

### 5. If a dictionary is stored in spam, what is the difference between the expressions &#39;cat&#39; in spam and&#39;cat&#39; in spam.keys()?

```python
spam = {'cat' : 'A hairy fluffy animal whose another name is a pussy xD'}
```

```python
'cat' in spam
```

```python
'cat' in spam.keys()
```

**_Both are similar.<br>
The expression, 'cat' in spam tries to check whether there is a key having 'cat' in the spam variable.<br>
The expression spam.keys(), retrieves every single key inside the spam variable._**

### 8. How do you &quot;pretty print&quot; dictionary values using which module and function?

**_The pprint module provides a capability to “pretty-print” arbitrary Python data structures in a form which can be used as input to the interpreter. If the formatted structures include objects which are not fundamental Python types, the representation may not be loadable._**

```python
import pprint
x = [{'d':[1,3,5],'c':[0,9,82,1],"a":[5,2,1,4]},
    {'d':[9,9,1],'c':[123,234,1],"a":[44,5,1,7]},
    {'d':[0,1,4,6],'c':[77,1,90,3],"a":[60,354,73]}]
```

```python
print(x)
```

```python
pprint.pprint(x)
```
