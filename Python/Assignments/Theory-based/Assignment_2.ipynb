### 5. What are the six comparison operators?

**Ans.**

_**Less than (<) — returns true if the value on the left is less than the value on the right, otherwise it returns false.<br>
Greater than (>) — returns true if the value on the left is greater than the value on the right, otherwise it returns false.<br>
Less than or equal to (<=) — returns true if the value on the left is less than or equal to the value on the right, otherwise it returns false.<br>
Greater than or equal to (>=) — returns true if the value on the left is greater than or equal to the value on the right, otherwise it returns false.<br>
Equal to (==) — returns true if the value on the left is equal to the value on the right, otherwise it returns false.<br>
Not equal to (!=) — returns true if the value on the left is not equal to the value on the right, otherwise it returns false.<br>**_

```python
print(5<6)
print(5>6)
print(5<=6)
print(5>=6)
print(5==6)
print(5!=6)
```

### 7. Recognize the following three blocks in this code:

Then finally the else block executes. Since, all the if conditions failed else block has to be executed.Here, it will print ham and spam and spam will be printed as it is outside the condtional statements.**_
