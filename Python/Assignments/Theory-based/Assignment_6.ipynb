### 1.What are escape characters, and how do you use them?

**_Escape characters generally represents the white spaces in a string. In other words to enter the illegal character in a string escape characters are used.<br>
It is used by a ('\\') backslash followed by a character which we want to use._**

**_1.\'	Single Quote	
2.\\	Backslash	
3.\n	New Line	
4.\r	Carriage Return	
5.\t	Tab	
6.\b	Backspace	
7.\f	Form Feed	
8.\ooo	Octal value	
9.\xhh	Hex value_**

### 2.What do the escape characters n and t stand for?

**_Escape character n: Stands for starting a new line after \n is being inserted._**

```python
print('Hello! \nLearning is fun.')
```

**_Escape character t: Stands for inserting tab spaces within the string._**

```python
print('Hello \tLearning \tis \tfun.')
```

### 9. What are the methods for right-justifying, left-justifying, and centering a string?

### 10. What is the best way to remove whitespace characters from the start or end?

**_To remove white spaces, strip function is used._**

```python
spam = '     Hey! Learning python is fun.           '
```

```python
spam
```

```python
print(spam.lstrip())
print(len(spam.lstrip()))
```

```python
print(spam.rstrip())
print(len(spam.rstrip()))
```

```python
print(spam.strip())
print(len(spam.strip()))
```
