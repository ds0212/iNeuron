### 17. How do you distinguish between copy.copy() and copy.deepcopy()?


_**Shallow copies duplicate as little as possible. A shallow copy of a collection is a copy of the collection structure, not the elements. With a shallow copy, two collections now share the individual elements.<br>
<br>
Deep copies duplicate everything. A deep copy of a collection is two collections with all of the elements in the original collection duplicated.**_

_**copy.deepcopy()<br>
deepcopy() creates new object and does real copying of original object to new one. Changing new deepcopied object doesn't affect original object.**_

```python
import copy
li1 = [1, 2, [3,5], 4] 
print("Original list: {}".format(li1))

result_A = [[90, 85, 82], [72, 88, 90]] 
result_B = copy.deepcopy(result_A) 
result_B[0][0] = 30

print("Original List: ")
print(result_A)
print("Deep Copy:")
print(result_B)
```

_**copy.copy()<br>
copy() create reference to original object. If we change copied object it also changes the original object.**_

```python
import copy

result_A = [[90, 85, 82], [72, 88, 90]] 
result_B = copy.copy(result_A) 
result_B[0][0] = 30

print("Original List: ")
print(result_A)
print("Shallow Copy:")
print(result_B)
```
