### 4.Write a Python Program to Check Armstrong Number?


```python
n = 371 #JUGAAD first time,Drawbacks: Valid only for 3 digit numbers.
a = []
cubedint = ''
b = str(n)
a.extend(b)
cubedint = cubedint + str((int(a[0])**3+int(a[1])**3+int(a[2])**3))
if int(cubedint) == n:
    print('It is an armstrong number.')
else:
    print('Not an armstrong number.')
```

```python
n = int(input('Enter the number to be checked: '))
r = 0
c = 0
sum = 0
temp = n
while n > 0:
    r = n % 10
    c = r * r * r
    sum = sum + c
    n = n // 10

n = temp  
if sum == n:
    print('Armstrong Number.')
else:
    print('Not an Armstrong Number.')
```

### 5.Write a Python Program to Find Armstrong Number in an Interval?

```python
upper = 1000
lower = 1
for num in range(lower , upper + 1):
    sum = 0
    temp = num
    
    while temp > 0:
        r = temp % 10
        c = r * r * r
        sum = sum + c
        temp = temp // 10
    
    if num == sum:
        print(num)
```
