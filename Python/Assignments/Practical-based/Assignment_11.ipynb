### 1.Write a Python program to find words which are greater than given length k?

```python
s = 'Hello! Thank you so much for giving me a wonderful oppurtunity. <3 '
ls = s.split()
output = []
n = 5
for i in ls:
    if len(i) > n:
        output.append(i)
```

### 4.Write a Python to check if a given string is binary string or not?

```python
s = '1010101010'
bs = set(s)
if '0' and '1' in bs:
    print('Binary')
else:
    print('Not Binary')
```

### 5.Write a Python program to find uncommon words from two Strings?

```python
A = "iNeuron in iNeuron" 
B = "Learning from iNeuron in iNeuron"
As = A.split()
Bs = B.split()
output = ''
for i in As:
    if i not in Bs:
        output = output + ' ' + i
for j in Bs:
    if j not in As:
        output = output + ' ' + j
output
```

### 6.Write a Python program to find all duplicate characters in string?

```python
s = 'Hello! Thank you so much for giving me a wonderful oppurtunity. <3'
```

```python
duplicated = []
for i in s:
    if s.count(i) > 1:
        duplicated.append(i)
print(set(duplicated))
```

### 7.Write a Python Program to check if a string contains any special character?

```python
s = 'Hello! Thank you so much for giving me a wonderful oppurtunity. <3 @ , \'"".'
from string import punctuation
punct = set(punctuation)
for i in s:
    if i in punct:
        print('Contains a special character.')
        break
else:
    print('Not a special character.')
```
