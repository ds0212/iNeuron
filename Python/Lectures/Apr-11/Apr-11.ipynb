```python
class student:
    def __init__(self, name, rollno, joining_date, current_topic):
        self.name = name
        self.rollno = rollno
        self.joining_date = joining_date
        self.current_topic = current_topic
        
    def nameParsing(self):
        if type(self.name) == list:
            for i in self.name:
                print('Name of a student', i)
        else:
            print('Not in a list.')
            
    def crt_topic(self):
        print('Current topic is:',self.current_topic)
        
    def str_rollno(self):
        try:
            if type(self.rollno) == type(str):
                print('Do nothing.')
            else:
                return str(self.rollno)
        except Exception as e:
            print('The error message', e)
        
    def duration(self, current_date):
        print('Duration of a student:', current_date - self.joining_date)
        
    def __str__(self):
        return "String class "
```

```python
kaps = student('kaps', 203, 2021, 'oops')
```

```python
kaps.duration(2022)
```

```python
kaps.str_rollno()
```

```python
print(kaps)
```

```python
class data:
    def __init__(self, file_name, file_type, date, size):
        self.file_name = file_name
        self.file_type = file_type
        self.data = date
        self.size = size
        
    def file_open(self):
        if not self.file_name in os.getcwd():
            f = open(self.file_name+'.txt', "w")
        f.write('Inserting some kind of data inside the file.')
        f.close()
        
    def file_read(self):
        try:
            f = open(self.file_name+'.txt', "r")
            print(f.read())
            f.close()
        except Exception as e:
            print('Error has occurred.', e)
            lg.exception(e)
        
    def file_append(self):
        f = open(self.file_name+'.txt', "a")
        f.write('Hello again!')
        f.close()
        
    def logging_module(self):
        lg.basicConfig(filename='app.log', level=lg.INFO ,format='%(asctime)s %(name)s %(levelname)s %(message)s', filemode = 'w')
        lg.info('Info added')
        lg.debug('Debug added')
        
```
