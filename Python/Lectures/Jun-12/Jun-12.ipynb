```python
a=[25, 36, 43, 67, 90, 45]
```

```python
def mergeSort(a):
    if len(a) > 1:
        mid = len(a) // 2
        l = a[:mid]
        r = a[mid:]
        mergeSort(l)
        mergeSort(r)

        i = j = k = 0

        while i < len(l) and j < len(r):
            if l[i] < r[j]:
                a[k] = l[i]
                i += 1
            else:
                a[k] = r[j]
                j += 1
            k += 1

            # Checking if any element was left
        while i < len(l):
            a[k] = l[i]
            i += 1
            k += 1

        while j < len(r):
            a[k] = r[j]
            j += 1
            k += 1
        return a
```

```python
print(mergeSort(a))
```

```python
s1 = 'pot'
s2 = 'Top'

def anagram():
    s = set(s1.lower())
    s_ = set(s2.lower())
    if s == s_:
        print('Yes')
    else:
        print('No')
```

```python
anagram()
```
