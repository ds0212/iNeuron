### Try:except block is  able to handle runtime error.

```python
try:
    a = open("test.txt" , "r")
except IOError:
    print('There is some issue with my code.')
```

```python
try:
    a = 5/0
except IOError:
    print('There is some issue with my code.')
```

```python
try:
    a = 5/0
except ArithmeticError as e:
    print('There is some issue with my code.',e)
```

```python
try:
    a = 4/4
except ArithmeticError as e:
    print('There is some issue with my code.',e)
else:
    print('This will execute once try will execute with success.')
    f = open('test.txt', 'r')
```

```python
try:
    a = 4/4
except ArithmeticError as e:
    print('There is some issue with my code.',e)
    f = open('test.txt', 'w')
else:
    print('This will execute once try will execute with success.')
    try:
        f = open('test.txt', 'r')
    except:
        print('Issue with internal try block.')
```

```python
def askforint():
    while True:
        try:
            a = int(input('Enter an interger:'))
        except Exception as e:
            print('This is the error msg', e)      
        else:
            print('Entered the correct value.')
            break
        finally:
            print('Close this issue.')
```
