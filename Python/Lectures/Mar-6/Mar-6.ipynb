# While:
#### To go from a one point to another point like a speedometer.
#### Keep on going till 100 with some conditions and could apply brakes or continue with the same thing again and again

```python
initial_speed = 0
final_speed =  100
count = 0
```

```python
while initial_speed < final_speed:
    print('speed')
    count = count + 1
    if count == 10:
        break
    
```
