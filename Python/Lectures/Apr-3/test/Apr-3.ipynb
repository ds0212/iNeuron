```python
def test5(n):
    for i in range(n):
        yield i ** 3
    
```

## Generator remembers what we have done last time, with the help of yield. It will keep on producing data without ending it.
i.e Displays the runtime output.

```python
for i in test5(30):
    print(i)
```

```python
a = 10
b = 36
for i in range(20):
    yield i
```

```python
def genfib(n):
    a = 0 
    b = 1
    output = []
    for i in range(n):
        yield a
        a,b = b,a+b
        
```

```python
genfib(6)
```

```python
for i in genfib(6):
    print(i)
```
