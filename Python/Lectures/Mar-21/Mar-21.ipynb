```python
def function1(*args):
    string = []
    integer = []
    dictionary = []
    tuples = []
    floats = []
    lists = []
    
    for i in args:
        if type(i) == str:
            string.append(i)
        elif type(i) == int:
            integer.append(i)
        elif type(i) == list:
            lists.append(i)    
        elif type(i) == dict:
            dictionary.append(i)
        elif type(i) == tuple:
            tuples.append(i)
        elif type(i) == float:
            floats.append(i)
    return string, integer, dictionary, tuples, floats,lists
                    
```

```python
function1(1,2,3,4,[4,5,6,6],{'a' : 'dhrumil', 'b':'sanghvi'},'sanghvi',45.0,(212,'dhru'),(222,212,'sang'),['ineuron','ai','9'],{'i':'iNeuron'},'dhrumil')
```

```python
def function2(*args,**kwargs):
    l = []
    for i in args:
        l.append(i)
    for j in kwargs:
        l.append(kwargs[j])
    return l
```

```python
function2([1,2,'dhru','sanghvi'],34324.00,['1','2','kapps',[1,2,3]],b = [4,5,6,7], c= [22,22,22,22,22])
```

```python
def fun1(*args):
    d = {}

    for i in args:
        print(type(i))
        if type(i) in d:
            d[type(i)].append(i)
        else:
            d[type(i)] = [i]
            
    return d
```
