# Querying techniques

```python
information.find_one()
```

```python
# SELECT * from information;
for rec in information.find():
    print(rec)
```

```python
for rec in information.find({}):
    print(rec)
```

```python
# Query the json documents based on equality conditions.
# SELECT * FROM information WHERE firstname='dhrumil';

for rec in information.find({'firstname':'Dhrumil'}):
    print(rec)
```

```python
# Query documents using query operators ($in, $lt, $gt)

for rec in information.find({'firstname':{'$in':['Dhrumil', 'Dhrumil2']}}):
    print(rec)
```

```python
# And and query operators.

for rec in information.find({'lastname':'Sanghvi2', 'age':{'$gt':3}}):
    print(rec)
```

```python
# OR operators.($or)

for rec in information.find({'$or':[{'firstname':'Dhrumil'},{'age' :11}]}):
    print(rec)
```

```python
# AND operators.($and)

for rec in information.find({'$and':[{'firstname':'Dhrumil2'},{'age' :11}]}):
    print(rec)
```
