# Abstraction - Private(___), protected('-'), public.

```python
class test:
    def __init__(self, a, b, c, d):
        self.__a = a
        self.b = b
        self.c = c
        self.d = d
    def test_custom(self, v):
        return v - self.__a
    
    def __str__(self):
        return 'Test code for abstraction.'
```

```python
o = test(3,4,5,6)
```

```python
o.test_custom(9)
```

```python
print(o)
```

```python
o.a
```

```python
o._a
```

```python
o.__a
```

```python
o._test__a
```

# Inheritance -  Allows to inherit the property from a parent class to a child class.

```python
class test1(test):
    def __init__(self, j, *args):
        super(test1, self).__init__(*args)
        self.j = j
        
m = test1(3,4,5,6,7)
m.b
m.j
```

```python
m.test_custom(9)
```

```python
class test:
    def __init__(self, a, b, c, d):
        self.__a = a
        self.b = b
        self.c = c
        self.d = d
    def test_custom(self, v):
        return v - self.__a

    def a(self):
        return 'A function in test class.'

    def __str__(self):
        return 'Test code for abstraction.'
```

```python
class test1(test):
    def __init__(self):
        super(test1,self).__init__()
    def a(self):
        return 'A function in test1 class.'
```

```python
class test2(test, test1):
    def __init__(self, j, *args):
        super(test2, self).__init__(*args)
        self.j = j
```

```python
o = test2(3,1,2,5,6)
```

```python
o.test1.a()
```

# Encapsulation - Extending the classes, trying to hide the information.

```python
class test:
    
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    
    def __str__(self):
        return "Test class."
    
class test1:
    
     def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    
     def __str__(self):
        return "Test1 class."
    
class test2:
    
     def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    
     def __str__(self):
        return "Test2 class."
    
class final:
    
    def __init__(self, x, y ,z):
        self.x = x
        self.y = y
        self.z = z
        
    def __str__(self):
        return  str(self.x) + " " + str(self.y) + " " + str(self.z)
```

```python
t = test(3,4,5)
t1 = test1(2,3,4)
t2 = test2(1,2,3)
f = final(t, t1, t2)
```
